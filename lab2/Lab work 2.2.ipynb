### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math
#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

def min_value(values_list):
    
    min_value = 0
    
    for value in values_list:
        if min_value > value:
            min_value = value
            
    print("Min value: ", min_value)
    return min_value

def max_value(values_list):
    
    max_value = 0
    
    for value in values_list:
        if max_value < value:
            max_value = value
            
    print("Max value: ", max_value)
    return max_value

def mean_value(values_list):
    
    mean_value = sum(values_list)/len(values_list)
        
    print("Mean value: ", mean_value)
    return mean_value

def median_value(values_list):
    if len(values_list) % 2 == 0:
        median_value = (values_list[int(len(values_list)/2) -1] + values_list[int(len(values_list)/2)])/2
    else:
        median_value = values_list[len(values_list)//2]
        
    print("Median: ", median_value)
    return median_value

def value_freq(values_list):
    
    set_with_random_numbers = set(values_list)
    dic = {}
    
    for value in set_with_random_numbers:
        dic[value] = values_list.count(value)
        
    print("Frequency of value: ", dic)
    return dic
    
def std_deviation(values_list):
    sum_delta = 0
    mean = mean_value(values_list)
    for value in values_list:
        sum_delta += (value - mean)**2
    
    sigma = math.sqrt(sum_delta/len(values_list))
    print("Standard deviation: ", sigma)
    return sigma
        

min_value(list_with_random_numbers)
max_value(list_with_random_numbers)
mean_value(list_with_random_numbers)
median_value(list_with_random_numbers)
std_deviation(list_with_random_numbers)
value_freq(list_with_random_numbers)


                    
```
