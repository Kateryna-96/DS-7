```python
info = df.describe()
info
```

```python
transport = df.groupby('Транспорт').count()
main_transport = transport[transport['ID']>200]
```

```python
df = pd.get_dummies(df, columns=['Транспорт'])
df
```

```python
transport.sort_values('ID', ascending=False)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_list = ['Маса', 'Відстань', 'Обєм']
for i in main_transport.index:
    x = f'Транспорт_{i}'
    x_list.append(x)

X = df[x_list]
y = df['Вартість']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_train
```

```python
reg_list = ['LinearRegression']
reg = LinearRegression()
reg.fit(X_train, y_train)
reg.coef_
```

```python
train_score = reg.score(X_train, y_train)
reg_list.append(train_score)
train_score
```

```python
test_score = reg.score(X_test, y_test)
reg_list.append(test_score)
test_score
```

```python
import sklearn.metrics as metrics
import math

y_pred = reg.predict(X_test)
mean_absolute_error=metrics.mean_absolute_error(y_test, y_pred) 
mse=metrics.mean_squared_error(y_test, y_pred) 
reg_list.append(mean_absolute_error)
reg_list.append(math.sqrt(mse))
print('MAE: ', mean_absolute_error)
print('MSE: ', mse)
print('RMSE: ', math.sqrt(mse))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg_list = ['RandomForestRegressor']
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
train_score = rf_reg.score(X_train, y_train)
rf_reg_list.append(train_score)
train_score
```

```python
test_score = rf_reg.score(X_test, y_test)
rf_reg_list.append(test_score)
test_score
```

```python
y_pred = rf_reg.predict(X_test)
mean_absolute_error=metrics.mean_absolute_error(y_test, y_pred) 
mse=metrics.mean_squared_error(y_test, y_pred) 
rf_reg_list.append(mean_absolute_error)
rf_reg_list.append(math.sqrt(mse))
print('MAE: ', mean_absolute_error)
print('MSE: ', mse)
print('RMSE: ', math.sqrt(mse))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr_list = ['DecisionTreeRegressor']
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
train_score = dtr.score(X_train, y_train)
dtr_list.append(train_score)
train_score
```

```python
test_score = dtr.score(X_test, y_test)
dtr_list.append(test_score)
test_score
```

```python
y_pred = dtr.predict(X_test)
mean_absolute_error=metrics.mean_absolute_error(y_test, y_pred) 
mse=metrics.mean_squared_error(y_test, y_pred) 
dtr_list.append(mean_absolute_error)
dtr_list.append(math.sqrt(mse))
print('MAE: ', mean_absolute_error)
print('MSE: ', mse)
print('RMSE: ', math.sqrt(mse))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr_list = ['GradientBoostingRegressor']
gbr = GradientBoostingRegressor(n_estimators=1000)
gbr.fit(X_train, y_train)
train_score = gbr.score(X_train, y_train)
gbr_list.append(train_score)
train_score
```

```python
test_score = gbr.score(X_test, y_test)
gbr_list.append(test_score)
test_score
```

```python
y_pred = gbr.predict(X_test)
mean_absolute_error=metrics.mean_absolute_error(y_test, y_pred) 
mse=metrics.mean_squared_error(y_test, y_pred) 
gbr_list.append(mean_absolute_error)
gbr_list.append(math.sqrt(mse))
print('MAE: ', mean_absolute_error)
print('MSE: ', mse)
print('RMSE: ', math.sqrt(mse))
```

```python
from sklearn.ensemble import VotingRegressor
voit_list = ['VotingRegressor']
dtr = DecisionTreeRegressor()
rf_reg = RandomForestRegressor()
gbr = GradientBoostingRegressor(n_estimators=1000)
voit_reg = VotingRegressor([('dtr', dtr), ('rf_reg', rf_reg), ('gbr', gbr) ])
voit_reg.fit(X_train, y_train)
train_score = voit_reg.score(X_train, y_train)
voit_list.append(train_score)
```

```python
train_score
```

```python
test_score = voit_reg.score(X_test, y_test)
voit_list.append(test_score)
test_score
```

```python
y_pred = voit_reg.predict(X_test)
mean_absolute_error=metrics.mean_absolute_error(y_test, y_pred) 
mse=metrics.mean_squared_error(y_test, y_pred) 
voit_list.append(mean_absolute_error)
voit_list.append(math.sqrt(mse))
print('MAE: ', mean_absolute_error)
print('MSE: ', mse)
print('RMSE: ', math.sqrt(mse))
```

```python
import numpy as np
index = np.array(['train_score', 'test_score', 'mean_absolute_error', 'RMSE'])
df_ml = pd.DataFrame(data={dtr_list[0]:dtr_list[1:], reg_list[0]:reg_list[1:], rf_reg_list[0]:rf_reg_list[1:], \
                     gbr_list[0]: gbr_list[1:], voit_list[0]:voit_list[1:]}, index=index)
df_ml
```
