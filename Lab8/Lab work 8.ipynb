```python
import pandas as pd

df = pd.read_csv('groceries.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

```python
from apyori import apriori

df1 = df.drop(columns=['Item(s)'])
```

```python
#перетворюємо у потрібний формат дані та опрацьовуємо за допомогою apriori
tranz_list = []
for i in range(9835):
    l = []
    for j in range(32):
        prod = str(df1.values[i,j])
        if prod!='nan':
            l.append(prod)
    tranz_list.append(l)
tranz_list
```

```python
result = list(apriori(tranz_list, max_length = 1, min_support=0.003, min_confidence = 0.1, min_lift = 1))
df_results = pd.DataFrame(result)
print('Топ-3 продукти:')
df_results.sort_values('support', ascending=False).head(3)
```

```python
result = list(apriori(tranz_list, max_length = 2, min_support=0.03, min_confidence = 0.1, min_lift = 1))
df_results = pd.DataFrame(result)
print('Топ-2 пари продуктів:')
df_results.sort_values('support', ascending=False, inplace=True)
top2 = [df_results.values[8], df_results.values[9]]
top2
```

Продукти, які продаються найкраще: цільне молоко, овочі та булочки. Причому найкраще продаються у парі цільне молоко з овочами та цільне молоко з булочками. Скоріше всього, варто ці продукти розміщати поряд для зручності покупців і збільшення confinence.

Завантажте файл https://www.kaggle.com/vjchoudhary7/customer-segmentation-tutorial-in-python
 
- Спробуйте провести кластеризацію, тобто виділити цільові групи, для яких можна впроваджувати певні маркетингові акції
- Опишіть отримані результати

```python
import pandas as pd
df = pd.read_csv('Mall_Customers.csv')
df = df.set_index('CustomerID')
df['Gender'] = df['Gender'].map(lambda x: 0 if x=='Male' else 1 )
df
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters=2)
```

```python
y_pred = cluster.fit_predict(df)
y_pred
```

```python
import matplotlib as plt
import matplotlib.pyplot
```

```python
fig, ax = plt.pyplot.subplots()
ax.scatter(df['Annual Income (k$)'], df[ 'Spending Score (1-100)'], c= y_pred)
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score (1-100)')
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster = DBSCAN(eps=2)
```

```python
y_pred = cluster.fit_predict(df)
```

```python
fig, ax = plt.pyplot.subplots()
ax.scatter(df['Annual Income (k$)'], df[ 'Spending Score (1-100)'], c= y_pred)
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score (1-100)')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=5)
```

```python
y_pred = cluster.fit_predict(df)
```

```python
fig, ax = plt.pyplot.subplots()
ax.scatter(df['Annual Income (k$)'], df[ 'Spending Score (1-100)'], c= y_pred)
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score (1-100)')
```
