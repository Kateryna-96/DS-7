### Завдання:

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
fig, ax = plt.subplots(2, 2, figsize=(15,8))
ax[0][0].plot(df['a'])
ax[0][1].plot(df['b'])
ax[1][0].plot(df['c'])
ax[1][1].plot(df.index)
```

Створити візуалізацію, аналогічно рисунку 

```python
fig, ax = plt.subplots(figsize = (30,15))
ax.plot(df.index, df["a"], linewidth=3)
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, ax = plt.subplots(2,  figsize = (20,10))
ax[0].plot(df['a'])
ax[1].plot(df['b'])
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
fig, ax= plt.subplots(figsize =(20, 10), linewidth = 5)

ax.plot(df.index, df['a'], df.index, df['b'], 'darkorange', linewidth = 2)
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
fig, ax= plt.subplots(figsize =(20, 10))
ax.plot(df.index, df['a'], df.index, df['b'], 'darkorange', linewidth = 2)
sns.set_style('darkgrid')
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
fig, ax= plt.subplots(figsize =(20, 10))
ax.plot(df.index, df['a'],'r--')
ax.plot(df.index, df['b'], 'darkorange', linewidth = 10)
ax.plot(df.index, df['c'], 'yellow', marker = 'o', linewidth = 1)
sns.set_style('darkgrid')
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
fig, ax = plt.subplots(3,  figsize = (20,10))
ax[0].plot(df['a'])
ax[0].legend(['Line A'], loc = 'lower right')
ax[1].plot(df['b'])
ax[1].legend(['Line B'], loc = 'center left')
ax[2].plot(df['c'])
sns.set_style('darkgrid')
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
fig, ax = plt.subplots(figsize = (20,10))
ax.bar(height=df['a'], x = df.index, color = 'tomato', alpha = 0.5, width=0.8)
ax.legend('a')
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
fig, ax = plt.subplots(figsize = (20,9))
ax.plot(df.index, df['a'], 'v-')
ax.set_xlabel('This is the X Axis', fontsize=20)
ax.yaxis.set_label_text('This is the Y Axis', fontsize=20)
ax.set_title('This is the Title', fontsize=20)
sns.set_style('darkgrid')
```
